```python
estados = [
    "AC",  "AL",  "AP",  "AM",  
    "BA",  "CE",  "DF",   "ES",   "GO",  
    "MA",  "MT",   "MS",   #"MG",  
    "PA",  "PB",  #"PR", 
    "PE",  "PI",  "RJ",   "RN",   "RS",   
    "RO",  "RR",  "SC",   
    #"SP",  
    "SE",  "TO"   
]
anos = [1997, 1998, 1999, 2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007,2008, 2009, 2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023, 2024
        ]
```

```python
Grupos = ['PA', 'BI', 'PS', 'SAD']

for Grupo in Grupos:
    try:
        # Obtém os arquivos para o grupo atual
        print(f"Processando o grupo: {Grupo}")
        files = sia.get_files([Grupo], uf=estados, year=anos)
        # Faz o download dos arquivos para o diretório especificado
        local_dir = f"pysus_sih/parquet_files/SIA/{Grupo}"
        parquet = sia.download(files, local_dir=local_dir)
        
        print(f"Arquivos para o grupo '{Grupo}' baixados com sucesso em: {local_dir}")
    except Exception as e:
        print(f"Erro ao processar o grupo '{Grupo}': {e}")
```
